```python
import pandas as pd
import json

# json_path = 'eval_scripts/summary_all_ori.json'
# json_path = 'eval_scripts/summary_all.json'
# json_path = 'eval_scripts/summary_2345_ori.json'
# json_path = 'eval_scripts/summary_2345.json'
# json_path = 'eval_scripts/summary_model_type_avg_score_round2345_ori.json'
json_path = 'eval_scripts/summary_model_type_avg_score_round2345_div.json'

with open(json_path, "r", encoding="utf-8") as file:
    data = json.load(file)

formatted_data = []
for model, types in data.items():
    for type, score in types.items():
        formatted_data.append({"Model": model, "Type": type, "Score": score['score']})

# 创建 DataFrame 并进行透视
df = pd.DataFrame(formatted_data)
pivot_df = df.pivot(index="Model", columns="Type", values="Score")
pivot_df
```

```python
# 设置指定的模型顺序
model_order = [
    "hive",
    "instruct-diffusion",
    "instructpix2pix",
    "magicbrush",
    "mgie",
    "iedit",
    "any2pix",
    "hqedit",
    # "EXP_MIN",
    # "EXP_MAX"
]
pivot_df = pivot_df.reindex(model_order)

# 设置指定的列顺序
column_order = [
    "animal",
    "object",
    "scenery",
    "plant",
    "human",
    "global",
]
pivot_df = pivot_df[column_order]
pivot_df
```

```python
def format_score(value, column):
    # if column == "StyleAlteration" or pd.isna(value):  # StyleAlteration
    #     return round(value, 2)
    return round(value * 100, 2)

pivot_df = pivot_df.apply(
    lambda col: col.apply(lambda val: format_score(val, col.name))
)
```
